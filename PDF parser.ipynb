```python
#TABULA

import tabula

import pandas as pd
```

```python
# опции guess и lattice - могут влиять на качество распознавания. 
for i in range (135, 142):
    df = tabula.read_pdf("electro.pdf", pages=i, guess = False, lattice = True)
    df.iloc[:,3:]=df.iloc[:,3:].apply(lambda x: x.str.replace(',','.')).astype(float)

    #df.iloc[:,10] = ''
    df.to_excel('output'+str(i)+'.xlsx')
```

```python
# эта часть кода дает возможность конвертировать элементы датафрейма из строк в числа "на лету", не сжимая размерность
df.iloc[:,3:]=df.iloc[:,3:].apply(lambda x: x.str.replace(',','.')).astype(float)
```

```python
df.iloc[:,3:10] = df.iloc[:,3:10].str.replace(',','.')
```
