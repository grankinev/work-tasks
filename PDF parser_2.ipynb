```python
import tabula
df = tabula.read_pdf("brazil.pdf", pages='all')
```

```python
df
```

```python
import pandas as pd
```

```python
years = pd.DataFrame(df.iloc[:40,2]).copy()
```

```python
years.drop(index=[0,1], inplace=True)
years.reset_index(drop = True, inplace=True)
years
```

```python
years1 = pd.DataFrame(df.iloc[:40,5]).copy()
years1.drop(index=[0,1], inplace=True)
years1.reset_index(drop = True, inplace=True)
years1.rename(columns={'real GDP indexes.1':'real GDP indexes'}, inplace=True)
years1
```

```python
years2 = pd.DataFrame(df.iloc[41:81,2]).copy()
years2.drop(index=[41,42], inplace=True)
years2.reset_index(drop = True, inplace=True)
years2
```

```python
years3 = pd.DataFrame(df.iloc[41:81,5]).copy()
years3.drop(index=[41,42], inplace=True)
years3.reset_index(drop = True, inplace=True)
years3.rename(columns={'real GDP indexes.1':'real GDP indexes'}, inplace=True)
```

```python
ally = pd.concat([years,years1,years2,years3], axis = 0, ignore_index=True)
```

```python
ally
```

```python
ally.to_excel("output.xlsx",sheet_name='Sheet_name_1')
```
